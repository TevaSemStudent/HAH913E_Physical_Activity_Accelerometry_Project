# Load the Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import time
import pandas as pd
```

# Read data from the next file : 23_10_25_raw_cut

```python
# Noms des fichiers dans le dossier data
files = [
    'data/24_11_raw.csv',
    'data/23_10_25_raw_cut.csv'
]

# Lecture et concat√©nation
df_list = []
for f in files:
    tmp = pd.read_csv(f, delimiter=',', skiprows=0)
    tmp.columns = ['t', 'x', 'y', 'z']
    df_list.append(tmp)

data = pd.concat(df_list, ignore_index=True)

print(data.shape)
print(data.head())
print(data.info())

# Conversion
data['t'] = pd.to_datetime(data['t'], format='%Y-%m-%d %H:%M:%S.%f')
data['x'] = pd.to_numeric(data['x'], errors='coerce', downcast='float')
data['y'] = pd.to_numeric(data['y'], errors='coerce', downcast='float')
data['z'] = pd.to_numeric(data['z'], errors='coerce', downcast='float')

g = 9.81
data['x'] /= 256
data['y'] /= 256
data['z'] /= 256

print('Conversion terminee')
print(data.head())
print(data.info())
```

```python
acceleration = np.sqrt(data['x']**2 + data['y']**2 + data['z']**2) - 1

# Affichage des donn√©es
plt.figure(figsize=(10, 4))
plt.plot(data['t'], -data['x'], color='gray', alpha=0.25, label='x component')
plt.plot(data['t'], -data['y']+10, color='gray', alpha=0.5, label='y component')
plt.plot(data['t'], -data['z']+20, color='gray', alpha=0.75, label='z component')
plt.plot(data['t'], acceleration+30, color='k', label='acceleration magnitude')
#plt.plot(data['t'], data['x'] - data['y'], color='b')
plt.xlabel('Time (hms)')
plt.ylabel('Acceleration (g)')
#plt.xlim([0, 20000])
#plt.ylim([-10, -6])
plt.legend()
plt.show()
```
